#### üîé Legal Conditions For Abortion
As per [¬ß218 Strafgesetzbuch (StGB)](https://www.gesetze-im-internet.de/stgb/BJNR001270871.html#BJNR001270871BJNG005302307), abortions are part of the German criminal law under offences against life and incur a penalty of imprisonment or fine.

[¬ß218a](https://www.gesetze-im-internet.de/stgb/__218a.html) excemptions are:
1. **counseling regulation:** when the pregnant woman requests the abortion and has sought counseling, the abortion is conducted by a physician, and no more than 12w have passed since conception 

2. **medical indication:** when the health of the pregnant woman is endangered and the danger cannot be averted in any other reasonable way

3. **criminological indication:** when the pregnancy is consequence of rape or sexual abuse

4. like (1) with a 10w extension but only the pregnant woman does not incur penalty (meaning to go  abroad) and if the pregnant woman was in exceptional distress at the time of the procedure
